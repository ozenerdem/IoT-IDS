```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import LocalOutlierFactor as LOF
from time import process_time
```

# <font color=black>Local Outlier Factor</font>

```python
df = pd.read_csv('dataset_maxabs.csv')
dataset = df.select_dtypes(include = ['float64', 'int64'])
dataset.head()
```

```python
#shuffling dataset
df1 = dataset.iloc[np.random.permutation(len(df))]
df2 = df1.iloc[np.random.permutation(len(df1))]
df3 = df2.iloc[np.random.permutation(len(df2))]
df3.head()
```

```python
df3 = df3.values
x = df3[:,:-1]
y = df3[:,[-1]]
df3 = np.array(df3)
```

```python
#masking outliers
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
print(x.shape, y.shape)
lof = LOF()
yhat = lof.fit_predict(x)
mask = yhat != -1
x_train, y_train = x[mask, :], y[mask]
print(x_train.shape, y_train.shape)
```

```python
df_x = pd.DataFrame(x_train)
df_y = pd.DataFrame(y_train)
dataset_lof = pd.concat([df_x, df_y], axis = 1)
```

```python
dataset_lof.to_csv("dataset_lof.csv", index = False)
```
