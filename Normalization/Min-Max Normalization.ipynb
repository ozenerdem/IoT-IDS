# <font color=black>Min-Max Normalization</font>

```python
dataset_x = dataset.iloc[:,:-4]
dataset_y = dataset.iloc[:,[-1]]
df = pd.concat([dataset_x, dataset_y], axis=1)
df.head()
```

```python
t1_start = process_time() 
ss = MinMaxScaler()
minmax_df = ss.fit_transform(dataset_x)
minmax_df = pd.DataFrame(minmax_df, columns = col_names)
t1_stop = process_time()
print("time:", t1_stop-t1_start)
```
